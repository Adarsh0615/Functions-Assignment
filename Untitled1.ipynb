#### 1. What is the difference between a function and a method in Python?

Ans :- A **function** is a block of code defined using the `def` keyword that performs a specific task and is independent of objects. A **method**, on the other hand, is a function that is associated with an object and is called on that object. Methods can access and modify the object’s attributes.

**Example:**
```python
# Function
def greet(name):
    return f"Hello, {name}!"

# Method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

# Usage
print(greet("Alice"))  # Function call
p = Person("Alice")
print(p.greet())       # Method call
```

#### 5. What are iterators in Python and how do they differ from iterables?
 Ans:- An iterator is an object that represents a stream of data and provides methods like iter() and next__(). An iterable is an object that can be looped over using a `for` loop, but it must be converted to an iterator to fetch data sequentially.

**Example:**

#### 6. Explain the concept of generators in Python and how they are defined.

Ans:- Generators are a special type of iterator that yield values one at a time using the `yield` keyword. They are defined like functions but use `yield` instead of `return`.

**Example:**
```python
def generate_numbers():
    for i in range(3):
        yield i

for number in generate_numbers():
    print(number)
```

#### 7. What are the advantages of using generators over regular functions?

Ans :- Generators are memory-efficient since they produce values on-the-fly.
- They allow lazy evaluation, which is useful for handling large datasets.
- Easier to implement than manually creating iterators.

**Example:**
```python
# Generator for infinite sequence
import itertools
infinite_gen = itertools.count(1)
for i in infinite_gen:
    if i > 5:
        break
    print(i)
```

#### 9. Explain the purpose and usage of the `map()` function in Python.

Ans :- The `map()` function applies a given function to all items in an iterable.

**Example:**
```python

```

#### 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans:- **`map()`**: Applies a function to all items in an iterable.
- **`reduce()`**: Applies a function cumulatively to reduce an iterable to a single value.
- **`filter()`**: Filters items in an iterable based on a condition.

**Example:**
```python
from functools import reduce
numbers = [1, 2, 3, 4]

# map()
squared = list(map(lambda x: x**2, numbers))
print(squared)  # Output: [1, 4, 9, 16]

# reduce()
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24

# filter()
even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)  # Output: [2, 4]
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

Ans:-

````python
def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage:
numbers = [1, 2, 3, 4, 5, 6]
print(sum_of_evens(numbers))
````

4. Write a Python function that checks if a given number is prime or not from 1 to 200.


Ans:-

````python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage:
print(is_prime(17))
print(is_prime(100))
````

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

Ans :-

````python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.next = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        value = self.current
        self.current, self.next = self.next, self.current + self.next
        self.count += 1
        return value

# Example usage:
fib = Fibonacci(10)
for num in fib:
    print(num, end=" ")
````

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


Ans :-

````python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_two(5):
    print(power, end=" ")
````

7. Implement a generator function that reads a file line by line and yields each line as a string.

Ans :-

````python
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
            print(line)
````

9. Write a Python program that uses map to convert a list of temperatures from Celsius to Fahrenheit.

Ans :-

````python
def celsius_to_fahrenheit(celsius):
    return list(map(lambda c: (c * 9/5) + 32, celsius))

# Example usage:
temperatures = [0, 20, 30, 40]
print(celsius_to_fahrenheit(temperatures))
````

10. Create a Python program that uses filter() to remove all the vowels from a given string.

Ans :-

````python
def remove_vowels(string):
    return ''.join(filter(lambda char: char.lower() not in 'aeiou', string))

# Example usage:
string = "beautiful"
print(remove_vowels(string))
````
